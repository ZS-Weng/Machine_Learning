## Load libraries model and processor

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
import torch
import re
```

```python
from transformers import DonutProcessor, VisionEncoderDecoderModel

processor = DonutProcessor.from_pretrained("naver-clova-ix/donut-base-finetuned-docvqa")
model = VisionEncoderDecoderModel.from_pretrained("naver-clova-ix/donut-base-finetuned-docvqa")
```

```python
def get_answer(folder_path, filename, question, model, processor):

    image = Image.open(folder_path/filename).convert('RGB')

    pixel_values = processor(image, return_tensors="pt").pixel_values
    
    prompt = f"<s_docvqa><s_question>{question}</s_question><s_answer>"
    decoder_input_ids = processor.tokenizer(prompt, add_special_tokens=False, return_tensors="pt")["input_ids"]

    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)

    outputs = model.generate(pixel_values.to(device),
                               decoder_input_ids=decoder_input_ids.to(device),
                               max_length=model.decoder.config.max_position_embeddings,
                               early_stopping=True,
                               pad_token_id=processor.tokenizer.pad_token_id,
                               eos_token_id=processor.tokenizer.eos_token_id,
                               use_cache=True,
                               num_beams=3,
                               bad_words_ids=[[processor.tokenizer.unk_token_id]],
                               return_dict_in_generate=True,
                               output_scores=True)
    
    seq = processor.batch_decode(outputs.sequences)[0]
    seq = seq.replace(processor.tokenizer.eos_token, "").replace(processor.tokenizer.pad_token, "")
    seq = re.sub(r"<.*?>", "", seq, count=1).strip()  # remove first task start token
    
    return processor.token2json(seq), outputs['sequences_scores'].item()*-1
```

```python
def show_image(image):
    img = image.resize((600,600))
    img = np.asarray(img)
    plt.figure(figsize=(7,7))
    plt.imshow(img)
```

## Looping through multiple images to extract information

```python
image_folder = Path("./Images")

#Creating Plot related details
fig = plt.figure(figsize=(12, 10))
rows = 2
columns = 4

for i,file in enumerate(image_folder.glob('*Invoice*')):
    
    # Adds a subplot at the 1st position
    fig.add_subplot(rows, columns, i+1)

    # showing image
    image = Image.open(image_folder/file.name)
    plt.imshow(image)
    plt.axis('off')
```

```python
image_folder = Path("./Images")

questions = ['What is the total or balance?',
             'What is the title at the top?']

details_dict = {'Image File':[],
                'Question':[],
                'Data':[]}



for question in questions:

    for file in image_folder.glob('*Invoice*'):

        answer, score = get_answer(image_folder, filename, question, model, processor)

        details_dict['Image File'].append(file.name)
        details_dict['Question'].append(answer['question'])
        details_dict['Data'].append(answer['answer'])
```

```python
df_results = pd.DataFrame(details_dict)
```

```python
df_results
```

```python
df_results_pivot = df_results.pivot_table(index='Image File',
                                          columns='Question',
                                          values='Data',
                                          aggfunc=lambda x: ''.join(str(v) for v in x)).reset_index()
```
